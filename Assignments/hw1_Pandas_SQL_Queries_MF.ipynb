```python
#Question 6
#6) Combine the table project and abstract and keep the ones that match for year 2013
# Query
# Define the query
QUERY_P = '''
SELECT *
FROM projects;
'''
QUERY_A = '''
SELECT *
FROM abstracts;
'''
#Create dataframes
Projects=pd.read_sql_query(QUERY_P,con=engine)
Abstracts=pd.read_sql_query(QUERY_A,con=engine)
#Merge
merge_q6 = pd.merge(Projects, Abstracts, on=["project_id"], how="inner")
#Filter for 2013 only
q6=merge_q6[(merge_q6['fy'] == 2013)]
# Plot
q6
```
